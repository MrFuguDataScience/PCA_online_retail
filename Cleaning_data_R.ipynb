```r
# Wholesale Retail from an Online Store:

onlineretail<-read.csv("online_sales.csv")
```

```r
nrow(onlineretail) # Row count
length(onlineretail) # Column count
head(onlineretail)
tail(onlineretail)
```

```r
# Number of Duplicate Rows:

dup_rows <-sum(duplicated(onlineretail))

dup_rows
```

```r
# Proportion of Data Left after removing duplicates:

round(1-dup_rows/nrow(onlineretail),digits=4)
```

```r
# Specific Rows that are duplicates:

spec_dup_rows<-onlineretail[duplicated(onlineretail),]
head(spec_dup_rows)
```

```r
# Not duplicated Rows, using the logical NOT (!):

onlineretail[!duplicated(onlineretail),]
```

```r
# Removing Duplicate Rows: 
online_retail_am <- onlineretail %>% distinct()
online_retail_am
```

```r
# Proportion of data with NA values:

round(sum(is.na(online_retail_am))/nrow(online_retail_am),digits=4)
```

```r
apply(is.na(online_retail_am), 2, which)
```

```r
?apply
# ?filter_all
```

```r
# Think (UNION) when using any_vars()
online_retail_am %>% filter_all(any_vars(is.na(.))) 
```

```r
# remove_na

remove_na<-na.omit(online_retail_am)
head(remove_na)
```

```r
# Proportion of data from origianl dataset left:
# round(nrow(remove_na)/nrow(onlineretail),digits=4)
# Amount of NA values from dataframe
round(1-nrow(remove_na)/nrow(online_retail_am),digits=4)

round(nrow(remove_na)/nrow(online_retail_am),digits=4)
```

```r
# Unique Countries Placing Orders:
unique(remove_na$Country)
```
